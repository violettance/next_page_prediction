```python
import pandas as pd
```

```python
df = pd.read_csv("bq_data_export.csv", low_memory=False)
```

```python
# Convert the "event_timestamp" column to numeric
df['event_timestamp'] = pd.to_numeric(df['event_timestamp'], errors='coerce')
```

```python
# Calculate the time duration and create a new column
df['duration'] = df['event_timestamp'].shift(-1) - df['event_timestamp']
```

```python
# Round the "duration" values to one decimal place and store them in a new column
df['readable_duration'] = df['duration'].apply(lambda x: round(x / 1e6, 1))
```
